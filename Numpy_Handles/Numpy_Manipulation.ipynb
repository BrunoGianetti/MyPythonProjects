### Adendo: Cópia ou Referência de arrays?

```python
# Cópias referenciadas: alteração em uma, altera outra
preco_imoveis = np.array([10000,120000,11000,200000])
preco_imoveis_sao_paulo = preco_imoveis
```

```python
preco_imoveis[0] = 120000
```

```python
preco_imoveis_sao_paulo
```

```python
# Cópias não referenciadas
preco_imoveis = np.array([10000,120000,11000,200000])
preco_imoveis_sao_paulo = np.copy(preco_imoveis) # <---
preco_imoveis[0] = 120000
preco_imoveis_sao_paulo
```

### exercício de plotagem de círculo

```python
# Gerar uma sequência de valores de x de -1 a 1
x = np.arange(-1, 1, 0.0001)

# Implementação da fórmula
y1 = np.sqrt(1 - x**2)
y2 = -np.sqrt(1 - x**2)

# Plotar o gráfico com as duas partes do círculo
plt.plot(x, y1, 'r')
plt.plot(x, y2, 'r' )

# Adicionar o título do gráfico e os rótulos dos eixos x e y
plt.title("Círculo")
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")

# Exibir o gráfico
plt.show()
```

### continuando com regressão liner

```python
# plotagem da regressão linear calculada corretamente
Y = Moscow
x = datas
n = np.size(Moscow)

(x**2).shape

a = (n*np.sum(x*Y) - np.sum(x)*np.sum(Y)) / (n*np.sum(x**2) - np.sum(x)**2)

b = np.mean(Y) - a*np.mean(x)

y = a*x + b

plt.plot(datas, Moscow)
plt.plot(x, y)
```

### trabalhando com aleatoriedade

```python
# gerando números inteiros aleatórios
np.random.randint(low=40, high=100, size=100)
```

```python
# gerando números float aleatórios
np.random.uniform(low=0.10, high=0.90, size=100)
```

```python
# testando coeficiente angular com esses números aleatórios
coef_angulares = np.random.uniform(low=0.10, high=0.90, size=100)
```

```python
for i in range(100):
  print(np.linalg.norm(Moscow - (coef_angulares[i]*x+b)))
```

```python
# agregando esses valores a um array
norma = np.array([])

for i in range(100):
  norma = np.append(norma, np.linalg.norm(Moscow - (coef_angulares[i]*x+b)))

norma
```

```python
# testando o coeficiente angular em qualquer posição do array
coef_angulares[1]
```

```python
#reprodutibilidade - SEED
#travando a aleatoriedade gerada

np.random.seed(16)
np.random.uniform(low=0.10, high=0.90, size=100)
```

```python
# se reproduzir o código sem o seed na sequência, não haverá a trava de aleatoriedade.
np.random.uniform(low=0.10, high=0.90, size=100)
```
